```python
import numpy as np
import matplotlib.pyplot as plt
import mdtraj as md
```

```python
colvarA=np.loadtxt("Inputs/topolA/COLVAR1",skiprows=5)
colvarB=np.loadtxt("Inputs/topolB/COLVAR2",skiprows=5)
xtcA=md.load("Inputs/topolA/traj_comp.xtc", top="Inputs/topolA/confout.gro")
xtcB=md.load("Inputs/topolB/traj_comp.xtc",top="Inputs/topolB/confout.gro")
```

```python
print('check the shape_trajectory')
print('xtcA', xtcA.xyz.shape)
print('xtcB', xtcB.xyz.shape)

print('check the shape_COLVAR')
print('colvarA', colvarA.shape)
print('colvarB', colvarB.shape)

print('check the frame number_trajectory')
print('xtcA', xtcA.n_frames)
print('xtcB', xtcB.n_frames)

print('check the frame number_COLVAR')
print('colvarA', colvarA.shape[0])
print('colvarB', colvarB.shape[0])
```

```python
phi_colvarA = colvarA[:,1]
psi_colvarA = colvarA[:,2]
phi_xtcA = md.compute_phi(xtcA)[1]
psi_xtcA = md.compute_psi(xtcA)[1]
```

```python
color1 = '#00e4ff'
color2 = '#000065'

alpha1 = 0.5
alpha2 = 0.25

marker1 = 'o'
marker2 = 'o'

size1 = 5
size2 = 5

N=1

plt.scatter(phi_colvarA[::N], psi_colvarA[::N], color=color1, alpha=alpha1, marker=marker1, s=size1, label='COLVAR A')
plt.legend(loc='lower right', fontsize=12, frameon=False)
plt.scatter(phi_xtcA[::N], psi_xtcA[::N], color=color2, alpha=alpha2, marker=marker2, s=size2, label='xtc A')
plt.xlim(-np.pi, np.pi)
plt.ylim(-np.pi, np.pi)
plt.xlabel(r'$\phi$ (rad)', fontsize=12)
plt.ylabel(r'$\psi$ (rad)', fontsize=12)
```

```python
phi_colvarB = colvarB[:,1]
psi_colvarB = colvarB[:,2]
phi_xtcB = md.compute_phi(xtcB)[1]
psi_xtcB = md.compute_psi(xtcB)[1]
```

```python
color1 = '#00e4ff'
color2 = '#000065'

alpha1 = 0.5
alpha2 = 0.25

marker1 = 'o'
marker2 = 'o'

size1 = 5
size2 = 5

N=1

plt.scatter(phi_colvarB[::N], psi_colvarB[::N], color=color1, alpha=alpha1, marker=marker1, s=size1, label='COLVAR B')
plt.legend(loc='lower right', fontsize=12, frameon=False)
plt.scatter(phi_xtcB[::N], psi_xtcB[::N], color=color2, alpha=alpha2, marker=marker2, s=size2, label='xtc B')
plt.xlim(-np.pi, np.pi)
plt.ylim(-np.pi, np.pi)
plt.xlabel(r'$\phi$ (rad)', fontsize=12)
plt.ylabel(r'$\psi$ (rad)', fontsize=12)
```

```python
phi_colvarA_std = np.std(phi_colvarA)
psi_colvarA_std = np.std(psi_colvarA)
phi_xtcA_std = np.std(phi_xtcA)
psi_xtcA_std = np.std(psi_xtcA)

phi_colvarB_std = np.std(phi_colvarB)
psi_colvarB_std = np.std(psi_colvarB)
phi_xtcB_std = np.std(phi_xtcB)
psi_xtcB_std = np.std(psi_xtcB)
```

```python
print('A')
print('COLVAR: phi standard deviation = ', phi_colvarA_std, 'rad')
print('xtc: phi standard deviation = ', phi_xtcA_std, 'rad')
print('COLVAR: psi standard deviation = ', psi_colvarA_std, 'rad')
print('xtc: psi standard deviation = ', psi_xtcA_std, 'rad')

print('B')
print('COLVAR: phi standard deviation = ', phi_colvarB_std, 'rad')
print('xtc: phi standard deviation = ', phi_xtcB_std, 'rad')
print('COLVAR: psi standard deviation = ', psi_colvarB_std, 'rad')
print('xtc: psi standard deviation = ', psi_xtcB_std, 'rad')
```
