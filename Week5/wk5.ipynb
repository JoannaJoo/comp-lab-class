```python
traj=md.load_trr("Inputs/em.trr",top="Inputs/em.gro")
```

```python
import mdtraj
traj = mdtraj.load("Inputs/em.trr",top="Inputs/em.gro")
psi_indices, phi_indices = [6, 8, 14, 16], [4, 6, 8, 14]
angles = mdtraj.geometry.compute_dihedrals(traj, [phi_indices, psi_indices])
```

```python
from math import pi
plt.title('A free energy surface')
plt.scatter(angles[:, 0], angles[:, 1], marker='x', c=traj.time)
cbar = plt.colorbar()
cbar.set_label('Time [ps]')
plt.xlabel(r'$\Phi$ Angle [radians]')
plt.xlim(-pi, pi)
plt.ylabel(r'$\Psi$ Angle [radians]')
plt.ylim(-pi, pi)
```

```python
import matplotlib.pyplot as plt
plt.hist2d(angles[:,1],angles[:,0], bins=100)
plt.xlabel('Phi angle')
plt.ylabel('Psi angle')
```
